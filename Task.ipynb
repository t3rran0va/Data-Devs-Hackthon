```python
# Select all weekly columns
week_cols = [col for col in df.columns if "Week" in col]

# Melt into long format
df_long = pd.melt(df, id_vars=["Store", "Year"], value_vars=week_cols,
                  var_name="Week_Type", value_name="Value")

# Separate 'Week' and 'Type' (Sales or Transactions)
df_long[["Week_Number", "Type"]] = df_long["Week_Type"].str.extract(r'Week (\d+)_(\w+)')

# Convert week number to integer
df_long["Week_Number"] = df_long["Week_Number"].astype(int)

# Pivot so Sales and Transactions are separate columns
df_long = df_long.pivot_table(index=["Store", "Year", "Week_Number"],
                              columns="Type", values="Value").reset_index()

# Optional: flatten columns
df_long.columns.name = None

print(df_long)
```

```python
plt.figure(figsize=(25, 15))
correlation_matrix = df_long.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Variables')
plt.show()
```

```python
df_long = df_long.sort_values(["Store", "Year", "Week_Number"]).reset_index(drop=True)

# Create lag features for Sales and Transactions
lag_weeks = [1, 2, 3]  # example: 1-week, 2-week, 3-week lags

for lag in lag_weeks:
    df_long[f"Sales_Lag{lag}"] = df_long.groupby("Store")["Sales"].shift(lag)
    df_long[f"Transactions_Lag{lag}"] = df_long.groupby("Store")["Transactions"].shift(lag)

# Drop rows with NaN values created by lag
df_long = df_long.dropna().reset_index(drop=True)
```

```python
train = df_long[df_long["Year"] < 2]
test = df_long[df_long["Year"] == 2]

x_train = train.drop(columns=["Sales"])  # Features for predicting Sales
y_train = train["Sales"]

x_test = test.drop(columns=["Sales"])
y_test = test["Sales"]
```

```python
print("Train shape:", train.shape)
print("Test shape:", test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
print("MAE:", mae)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Assuming y_test and y_pred exist
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("RMSE:", rmse)
print("R²:", r2)
```
